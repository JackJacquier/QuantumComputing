```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere, plot_bloch_vector
from qiskit.visualization import plot_histogram
from qiskit_aer import Aer

import matplotlib.pylab as plt
import numpy as np
import pandas as pd
```

## One-qubit Hadamard

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 1000
results = backend.run(qc, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

## Two-qubit Hadamard

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 500000
results = backend.run(qc, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

## Generating Uniform distribution 

```python
nb_qubits = 8
qc = QuantumCircuit(nb_qubits)

for i in range(nb_qubits):
    qc.h(i)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 1000000
results = backend.run(qc, shots=shots).result()
output = results.get_counts()
plot_histogram(output, legend = None, bar_labels=None)
```
