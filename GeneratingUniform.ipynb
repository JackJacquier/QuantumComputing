```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere, plot_bloch_vector
from qiskit import Aer, execute
import matplotlib.pylab as plt
from qiskit.visualization import plot_histogram
import numpy as np
import pandas as pd
```

## One-qubit Hadamard

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 1000
results = execute(qc, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

## Two-qubit Hadamard

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 10000
results = execute(qc, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

## Generating Uniform distribution 

```python
n = 7
qc = QuantumCircuit(n)
for i in range(n):
    qc.h(i)
qc.measure_all()
qc.draw('mpl')

backend = Aer.get_backend('qasm_simulator')
shots = 10000
results = execute(qc, backend=backend, shots=shots).result()
output = results.get_counts()
plot_histogram(output, legend = None, bar_labels =None)
```
