# Comparison FFT / QFT speed

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def classical(N):
    return N*N

def dft(N):
    return N*np.log(N)

def qft(N):
    return np.log(N)**2


NN = np.arange(2, 100, 10)

#plt.plot(NN, classical(NN), label="classical")
plt.plot(NN, dft(NN), label="DFT")
plt.plot(NN, qft(NN), label="QFT")
plt.legend(loc="best")
plt.show()
```
